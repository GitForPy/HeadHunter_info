```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time

headers = {
    'Host': 'hh.ru',
    'User-Agent': 'Mozilla/5.0',
    'Accept': '*/*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive'
}

base_url = "https://hh.ru/vacancies/data-scientist?page="

jobs = []

for i in range(7):  # для 7 страниц
    print(f"Loading page {i+1}")
    url = base_url + str(i)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    job_listings = soup.find_all('div', class_='serp-item')

    for job in job_listings:
        job_title_tag = job.find('a', class_='serp-item__title')
        if job_title_tag:
            title = job_title_tag.text.strip()
            link = job_title_tag['href']
            salary_tag = job.find('span', attrs={'data-qa': 'vacancy-serp__vacancy-compensation'})
            salary = salary_tag.text.strip() if salary_tag else 'Не указано'
            company = job.find('a', attrs={'data-qa': 'vacancy-serp__vacancy-employer'})
            company = company.text.strip() if company else 'Не указано'
            jobs.append({'company': company, 'title': title, 'link': link, 'salary': salary})

    # добавляем задержку, чтобы не перегружать сервер запросами
    time.sleep(1)

df = pd.DataFrame(jobs)
```
